```python
#! /usr/bin/env python3
# -*- coding: utf-8 -*-

import matplotlib.pyplot as plt
y_axis=[]
van_eck=[0]
i=0
lim=int(input("Entrez le nombre de termes de la s√©quence : "))
while i<lim:
    y_axis.append(van_eck[-1])
    nb=van_eck.count(van_eck[-1])
    if nb==1: van_eck.append(0)
    else: van_eck.append(van_eck[-2::-1].index(van_eck[-1])+1)
    i+=1
x_axis=[n for n in range(len(y_axis))]
plt.plot(x_axis,y_axis)
plt.title("Nombres de Van Eck")
plt.xlabel("√âtape n de la s√©rie")
plt.ylabel("Nombres de Van Eck")
plt.show()
```
